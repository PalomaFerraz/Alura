```python
import pandas as pd 
import numpy as np
```

```python
size = 100
idade_max = 100
ids = np.arange(1, size+1, 1, dtype=int)
idades = np.random.randint(18, idade_max, size= size)
sinistros = np.random.randint(5, size= size)
premio = np.random.randint(500000, size= size)/100
sexo = np.random.choice(["F", "M"], size = size)
estado_civil = np.random.choice(["Solteiro", "Casado", "Divorciado", "Viuvo"], size = size)
```

```python
dict = { 'ids': ids,
         'idade': idades,
         'sinistros': sinistros,
         'premio': premio,
         'sexo': sexo,
         'est_civil': estado_civil}

dados = pd.DataFrame(dict)
dados.head()
```

```python
dados['categoria_num'] = np.digitize(dados['idade'], [31, 46])
choicelist = [ 'Jovem', 'Adulto', 'Senior']
condlist = [dados['categoria_num'] == 0, dados['categoria_num']== 1, dados['categoria_num']==2]
dados['categoria_desc1'] = np.select(condlist, choicelist, default= 'nan')

dados["categoria_desc2"] = pd.cut(dados["idade"], [18, 30, 45, idade_max], labels = ["Jovem", "Adulto", "Senior"], include_lowest = True)
dados
```

```python
desconto = pd.DataFrame({
'categoria_desc1': ['Jovem', 'Adulto', 'Senior'],
'tx_desconto': [0.05, 0.1, 0.1]
})

dados = pd.merge(dados, desconto, how='left', on='categoria_desc1')

dados['desconto'] = np.where(dados['sinistros'] == 0, dados['premio'] * dados['tx_desconto'], 0)
dados
```
