```python
import pandas as pd 
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
teste = pd.read_html("https://www2.susep.gov.br/menuestatistica/RankRoubo/resp_menu1.asp", header=0, thousands = '.', decimal = ',')
teste
```

```python
df = teste[0]
df.head()
```

```python
df.columns = ['Modelo', 'Indice_Roubo', 'Veiculos_Expostos', 'Sinistros']
df['Sinistros'] = df['Sinistros'].map(int)
df.head()

```

```python
df.sort_values('Indice_Roubo', axis=0, ascending=False, inplace=True)
df.head(10)
```

```python
df[['Veiculos_Expostos', 'Sinistros']].describe()
```

```python
df2 = df[(df['Veiculos_Expostos'] >= df['Veiculos_Expostos'].mean())]
df2.head(10)
```
