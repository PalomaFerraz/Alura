```python
import pandas as pd
import requests as rq
```

```python
r = rq.get('https://www.alura.com.br/api/cursos')
r.status_code
r.headers['content-type']


base = r.json()
base
dados = pd.json_normalize(data = base)
dados
```

```python
slug = dados['slug'].unique()

r2 = []

for i in slug:
    r2 += [pd.json_normalize(rq.get('https://cursos.alura.com.br/api/curso-{}'.format(i)).json(), max_level= 0)]

dados2 = pd.concat(r2)

dados2.head(15)
```

```python
dados2.drop(['ementa',
'chamadas',
'publico_alvo',
'video_1a_aula',
'instrutores',
'musicas',
'depoimentos',
'transcriptions',
'highlightedExtraField',
'curso_substituto',
'meta_title',
'video_1a_aula',
'parceiros',
'formacoes'], axis=1, inplace=True)
dados2.head(15)
```

```python
pd.set_option("display.max_columns",40)
dados2.head(15)
dados2.reset_index(inplace=True)
```

```python
dados2['alunos_ativos_total']=pd.json_normalize(dados2.alunos_ativos)['total']
dados2['subcategoria_nome']=pd.json_normalize(dados2.subcategoria)['nome']
dados2['categoria_nome']=pd.json_normalize(dados2.categoria)['nome']
dados2.head(15)
```

```python
dados2['quantidade_alunos'].fillna(0)
classes = ['Moderadamente Popular', 'Popular', 'Muito Popular']
bins = [dados2['quantidade_alunos'].min(), dados2['quantidade_alunos'].quantile(0.33), dados2['quantidade_alunos'].quantile(0.66), dados2['quantidade_alunos'].max()]

dados2['popularidade'] = pd.cut(x = dados2['quantidade_alunos'], bins = bins, labels= classes, include_lowest=True)

```

```python
teste = dados2.groupby(['popularidade'], as_index=False)[['popularidade']].size()
teste
```

```python
dados2 = dados2[dados2['nota_disponivel'] == True]
dados2.value_counts('nota_disponivel')
```

```python
nota_categoria = dados2.groupby(['categoria_nome'])['nota'].agg({'min', 'max', 'mean', 'std'}).round(1).reset_index()
nota_categoria.columns = ['Categoria_nome', 'Máximo', 'Mínimo', 'Média', 'Desvio Padrão']
nota_categoria
```

```python
nota_categoria2= nota_categoria.sort_values(by=["Média", "Máximo"])
nota_categoria2
```

```python
qtd_cursos = dados2.groupby(['categoria_nome', 'subcategoria_nome'], as_index=False)['subcategoria_nome'].size().rename(columns={'size': 'Quantidade'})
qtd_cursos['Percentual'] = (qtd_cursos.Quantidade / sum(qtd_cursos.Quantidade) * 100).round(2)
qtd_cursos
```
